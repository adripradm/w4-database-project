## QUERIES

```python

str_conn='mysql+pymysql://adri:password@localhost:3306/movies'


motor=create_engine(str_conn)
```

```python
pd.read_sql('''SELECT     film.title AS title,    category.category_id AS category_id,    category.name AS category,    CONCAT(hdd.first_name, ' ', hdd.last_name) AS actorname FROM    hdd        LEFT JOIN    film ON film.title = hdd.title        LEFT JOIN    category ON hdd.category_id =category.category_id GROUP BY hdd.title ORDER BY category.name''',motor)


#Query que agrupa peliculas por genero y te dice que actores aparecen en ella
```

```python
pd.read_sql('''select 

(film.rental_rate/film.replacement_cost) as rentabilidad,
film.title as title,
category.name as category


 from hdd       
 left join film
 on hdd.title=film.title
 
 left join category
 on hdd.category_id=category.category_id
 
 group by hdd.title
 order by rentabilidad desc
 limit 10
 ''',motor)
#Rentabilidad relativa por tittulo 
```

```python
pd.read_sql('''Select rental.inventory_id as id,
count(*) From   rental 
Group By  rental.inventory_id''',motor)

#No hay diferencia entre rentabilidad absoluta y relativa ya que la frecuencia es igual a 1 en todas la peliculas
```

```python
pd.read_sql('''Select 
concat(hdd.first_name, ' ', hdd.last_name) as actorname,

count(*) frecuencia

From   hdd
Group By  actorname

order by frecuencia desc
limit 10''',motor)

#Actores que mas salen en las peliculas 
```

```python
pd.read_sql('''select     

category,
count(category_id) as frecuency
 
 
from ( select 
 
category.name as category,
film.title as title,
rental.inventory_id as inventory_id,
inventory.film_id as film_id,
category.category_id as category_id
 
 


from hdd

left join category
on category.category_id=hdd.category_id

left join  film
on film.title=hdd.title

left join inventory
on film.film_id=inventory.film_id

left join  rental
on rental.inventory_id=inventory.inventory_id


Group By  film.title) as t

group by category

order by frecuency desc''',motor)

 #query para determinar cual es la categoria que mas se alquila
```

```python
pd.read_sql('''select 
customer_id,
count(customer_id) as peliculascliente
from rental

group by customer_id
order by count(customer_id) desc

limit 10''',motor)

#10 clientes que mas peliculas sacan
```

```python
pd.read_sql('''select               

category as categories,
count(category_id)*rentabilidad as rent
 
 
from ( select 
 (film.rental_rate/film.replacement_cost) as rentabilidad,
category.name as category,
film.title as title,
rental.inventory_id as inventory_id,
inventory.film_id as film_id,
category.category_id as category_id
 
 


from hdd

left join category
on category.category_id=hdd.category_id

left join  film
on film.title=hdd.title

left join inventory
on film.film_id=inventory.film_id

left join  rental
on rental.inventory_id=inventory.inventory_id


Group By  film.title) as t

group by category

order by rent desc''',motor)
#rentabilidad absoluta por generos 
```

```python
pd.read_sql('''SELECT 



datediff(cast(return_date as date), cast(rental_date as date)) as rental_duration,
category.name as category,
hdd.title as title

from hdd

left join category
on category.category_id=hdd.category_id

left join  film
on film.title=hdd.title

left join inventory
on film.film_id=inventory.film_id

left join  rental
on rental.inventory_id=inventory.inventory_id


group by title
order by rental_duration desc
limit 10''',motor)

#10 Peliculas que mas tardan en devolverse indicando al genero al que pertenecen
```

```python
pd.read_sql('''Select  
concat(hdd.first_name, ' ', hdd.last_name) as actorname,
category.name as category

From   hdd

left join category
on category.category_id=hdd.category_id

where hdd.first_name='penelope' 

Group By  category''',motor)


#categorias en las que aparece penelope guiness
```

```python
pd.read_sql('''Select   
film.rental_rate as rental_rate,
category.name as category,
count(film.title) as total


From   hdd

left join category
on category.category_id=hdd.category_id
left join  film
on film.title=hdd.title

left join inventory
on film.film_id=inventory.film_id

left join  rental
on rental.inventory_id=inventory.inventory_id
 
where rental_rate<4
Group By  category
order by total desc''',motor)


#numero de peliculas que cuesta menos de 4 por generos
```
